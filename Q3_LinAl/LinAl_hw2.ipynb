1. –£—Å—Ç–∞–Ω–æ–≤–∏—Ç—å, –∫–∞–∫–∏–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è –º–∞—Ç—Ä–∏—Ü  ùê¥ùêµ  –∏  ùêµùê¥  –æ–ø—Ä–µ–¥–µ–ª–µ–Ω—ã, –∏ –Ω–∞–π—Ç–∏ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –º–∞—Ç—Ä–∏—Ü:

–∞)  ùê¥  ‚Äî –º–∞—Ç—Ä–∏—Ü–∞  4√ó2 ,  ùêµ  ‚Äî –º–∞—Ç—Ä–∏—Ü–∞  4√ó2 ;

–ú–∞—Ç—Ä–∏—Ü—ã –ê –∏ –í –Ω–µ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–µ, –æ–¥–∏–Ω–∞–∫–æ–≤–æ–π —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–∏. –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ —É–º–Ω–æ–∂–∞—Ç—å –∏—Ö –º–æ–∂–Ω–æ, –µ—Å–ª–∏ 1 –∏–∑ –Ω–∏—Ö —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞—Ç—å. –ï—Å–ª–∏ —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞—Ç—å B, –ø–æ–ª—É—á–∏—Ç—Å—è –º–∞—Ç—Ä–∏—Ü–∞ 4*4 –ï—Å–ª–∏ —Ç—Ä–∞–Ω—Å–ø–æ–Ω–∏—Ä–æ–≤–∞—Ç—å A, —Ç–æ –ø–æ–ª—É—á–∏—Ç—Å—è –º–∞—Ç—Ä–∏—Ü–∞ 2*2

```python
A = np.random.randn(4, 2)
B = np.random.randn(4, 2)
print('–ú–∞—Ç—Ä–∏—Ü–∞ –ê:')
print(A)
print('–ú–∞—Ç—Ä–∏—Ü–∞ –í:')
print(B)
print('–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –ê * –í.–¢:')
print(A.dot(B.T))
print('–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –ê.–¢ * –í:')
print(B.T.dot(A))
```

–±)  ùê¥  ‚Äî –º–∞—Ç—Ä–∏—Ü–∞  2√ó5 ,  ùêµ  ‚Äî –º–∞—Ç—Ä–∏—Ü–∞  5√ó3 ;
–ú–∞—Ç—Ä–∏—Ü—ã ùê¥*ùêµ –∏–º–µ–µ—Ç —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å 2 3 
–ú–∞—Ç—Ä–∏—Ü–∞ –í*–ê –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç. 
–ú–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –º–∞—Ç—Ä–∏—Ü—É A.T * B.T. –ü–æ–ª—É—á–µ–Ω–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ –±—É–¥–µ—Ç –∏–º–µ—Ç—å —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å 3 * 2

```python
A = np.random.randn(2, 5)
B = np.random.randn(5, 3)
print('–ú–∞—Ç—Ä–∏—Ü–∞ –ê:')
print(A)
print('–ú–∞—Ç—Ä–∏—Ü–∞ –í:')
print(B)
print('–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –ê * –í:')
print(A.dot(B))
print('–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –í.–¢ * –ê.–¢:')
print(B.T.dot(A.T))
```

–≤)  ùê¥  ‚Äî –º–∞—Ç—Ä–∏—Ü–∞  8√ó3 ,  ùêµ  ‚Äî –º–∞—Ç—Ä–∏—Ü–∞  3√ó8 ;
–û–±–µ –º–∞—Ç—Ä–∏—Ü—ã ùê¥ùêµ –∏ ùêµùê¥ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω—ã 
–ú–∞—Ç—Ä–∏—Ü—ã ùê¥*ùêµ –∏–º–µ–µ—Ç —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å 8 * 8. 
–ú–∞—Ç—Ä–∏—Ü–∞ –í * –ê –∏–º–µ–µ—Ç —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å 3 * 3

```python
A = np.random.randn(8, 3)
B = np.random.randn(3, 8)
print('–ú–∞—Ç—Ä–∏—Ü–∞ –ê:')
print(A)
print('–ú–∞—Ç—Ä–∏—Ü–∞ –í:')
print(B)
print('–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –ê * –í:')
print(A.dot(B))
print('–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –í * –ê:')
print(B.dot(A))
```

–≥)  ùê¥  ‚Äî –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞  4√ó4 ,  ùêµ  ‚Äî –∫–≤–∞–¥—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞  4√ó4 .
–û–±–µ –º–∞—Ç—Ä–∏—Ü—ã  ùê¥*ùêµ  –∏  ùêµ*ùê¥  –æ–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –∏ –∏–º–µ—é—Ç —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å  4√ó4

```python
A = np.random.randn(4, 4)
B = np.random.randn(4, 4)
print('–ú–∞—Ç—Ä–∏—Ü–∞ –ê:')
print(A)
print('–ú–∞—Ç—Ä–∏—Ü–∞ –í:')
print(B)
print('–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –ê * –í:')
print(A.dot(B))
print('–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –í * –ê:')
print(B.dot(A))
```

2. –ù–∞–π—Ç–∏ —Å—É–º–º—É –∏ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü  ùê¥=([[1, -2], [3, 0]])  –∏  ùêµ=([[4, -1], [0, 5]]).

```python
A = np.array([[1, -2], [3, 0]])
B = np.array([[4, -1], [0, 5]])
print('–ú–∞—Ç—Ä–∏—Ü–∞ –ê:')
print(A)
print('–ú–∞—Ç—Ä–∏—Ü–∞ –í:')
print(B)
print('–°—É–º–º–∞ –º–∞—Ç—Ä–∏—Ü:')
print(A + B)
print('–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –ê * –í:')
print(A.dot(B))
```

3. –ò–∑ –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–µ–π —Å–ª–æ–∂–µ–Ω–∏—è –∏ —É–º–Ω–æ–∂–µ–Ω–∏—è –º–∞—Ç—Ä–∏—Ü –Ω–∞ —á–∏—Å–ª–æ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥, —á—Ç–æ –º–∞—Ç—Ä–∏—Ü—ã –æ–¥–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –æ–±—Ä–∞–∑—É—é—Ç –ª–∏–Ω–µ–π–Ω–æ–µ –ø—Ä–æ—Å—Ç—Ä–∞–Ω—Å—Ç–≤–æ. 
–í—ã—á–∏—Å–ª–∏—Ç—å –ª–∏–Ω–µ–π–Ω—É—é –∫–æ–º–±–∏–Ω–∞—Ü–∏—é  3ùê¥‚àí2ùêµ+4ùê∂  –¥–ª—è –º–∞—Ç—Ä–∏—Ü  
ùê¥=([[1, 7], [3, -6]]) ,  
ùêµ=([[0,5], [2, -1]]) ,  
ùê∂=([[2, -4], [1, 1]]).

#### 3ùê¥‚àí2ùêµ+4ùê∂ = [[3*1 - 0 + 4*2, 3*7 -2*5 - 4* 4],[3*3 - 2*2 +4, -6*3 +1*2 +4*1]] =
#### = [[11, -5],[9, 12]]

```python
#–ü—Ä–æ–≤–µ—Ä–∫–∞
A = np.array([[1, 7], [3, -6]])
B = np.array([[0,5], [2, -1]])
C = np.array([[2, -4], [1, 1]])
print('–ú–∞—Ç—Ä–∏—Ü–∞ –ê:')
print(A)
print('–ú–∞—Ç—Ä–∏—Ü–∞ –í:')
print(B)
print('–ú–∞—Ç—Ä–∏—Ü–∞ –°:')
print(C)
print('–†–µ–∑—É–ª—å—Ç–∞—Ç 3ùê¥‚àí2ùêµ+4ùê∂ :')
print(3*A - 2*B + 4*C)
```

4. –î–∞–Ω–∞ –º–∞—Ç—Ä–∏—Ü–∞  ùê¥=[[4, 1], [5, -2], [2, 3]] . –í—ã—á–∏—Å–ª–∏—Ç—å  ùê¥ùê¥ùëá  –∏  ùê¥ùëáùê¥ .

```python
A = np.array([[4, 1], [5, -2], [2, 3]])
print('–ú–∞—Ç—Ä–∏—Ü–∞ –ê:')
print(A)
print('–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –ê * –ê.–¢:')
print(np.dot(A, A.T))
print('–†–µ–∑—É–ª—å—Ç–∞—Ç —É–º–Ω–æ–∂–µ–Ω–∏—è –ê.–¢ * –ê:')
print(np.dot(A.T, A))
```
